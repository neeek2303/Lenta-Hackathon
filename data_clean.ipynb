## Drop very and less popular

```python
mat =  pd.DataFrame({'material':transactions.material}) 
```

```python
mat
```

```python
mat['number_of_rep'] = mat.material.map(mat.material.value_counts())
```

```python
mat = mat.drop_duplicates(keep='last')
```

```python
popularity = np.array( mat[(mat['number_of_rep'] < mat['number_of_rep'].quantile(populariti_max)) & (mat['number_of_rep'] > mat['number_of_rep'].quantile(0.00)) ]['number_of_rep'])
```

```python
sorted_p = np.sort(popularity)[::-1]
y = np.array([np.sum(sorted_p[:i]/np.sum(sorted_p)) for i in range(1, len(sorted_p)+1)])
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx], idx, sorted_p[idx]
```

```python
_, __, amount =  find_nearest(y, populariti_take) # 0.95 by default
_, __, amount
```

```python
mat_filtered = mat[mat['number_of_rep']>amount]
```
